```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from xgboost import XGBRegressor
```

```python
data=pd.read_csv('UBER_500.csv')
print(data)
```

```python
data.describe()
```

```python
x=data.drop(['PRICE','TIME_OF_DAY','TRAFFIC_LEVEL'],axis=1)
y=data['PRICE']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print('80% traning data is (rows)',xtrain.shape)
print('20% testing data is (rows)',xtest.shape)
print("80% training data is ",ytrain.shape)
print("80% training data is ",ytest.shape)
```

```python
xgb =XGBRegressor()
xgb.fit(xtrain,ytrain)
```

```python
output=xgb.predict(xtest)
print(output)
```

```python
score=r2_score(ytest,output)
abs(score)
```

```python
NCA=int(input("NO OF CARS AVAILABLE :"))
RAIN=int(input("IS IT RAINING IF YES(1) IF NOT (0):"))
DIST=int(input("DISTANCE IN KM:"))
user_data=[[NCA,RAIN,DIST]]
predictedprice=xgb.predict(user_data)
print("Predicted Price:",predictedprice)
```
