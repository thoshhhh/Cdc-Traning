```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
data=np.array([[1,6],[2,8],[5,12],[6,14],[9,24],[10,28]])
kmeans=KMeans(n_clusters=3,random_state=0)
kmeans.fit(data)
print(kmeans.labels_)
```

```python


plt.scatter(data[:,0],data[:,1],c=kmeans.labels_,s=75,cmap='rainbow')
plt.title('Kmeans clustering')
plt.xlabel('Experience-yrs')
plt.ylabel('Salary-lpa')
plt.show()
```

```python
import matplotlib.pyplot as ply
from sklearn.cluster import DBSCAN
import numpy as np
data=np.array([[1,6],[2,8],[5,12],[6,14],[9,24],[10,28]])
dbscan=DBSCAN(eps=3,min_samples=2)
dbscan.fit(data)
print(dbscan.labels_)
```

```python
plt.scatter(data[:,0],data[:,1],c=dbscan.labels_,s=75,cmap='rainbow')
plt.title('DBSCAN clustering')
plt.xlabel('Experience-yrs')
plt.ylabel('Salary-lpa')
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage,fcluster
labels = fcluster(linkage(data,method='ward'),t=3,criterion='maxclust')
print('Cluster lables : ',labels)
plt.scatter(data[:,0],data[:,1],c=labels,s=50,cmap='rainbow')
plt.title("Hierarchical clustering-Employees")
plt.xlabel('Experience-yrs')
plt.ylabel('Salary-lpa')
plt.show()
```
