```python
import seaborn as sns
import matplotlib.pyplot as plt
x=['sun','mon','tues','wed','thu','fri','sat']
y=[10,7,4,6,2,5,3]
sns.scatterplot(x=x,y=y,hue=x,style=x,s=100,palette="rocket",legend=True)
plt.xlabel("DAYS")
plt.ylabel("AMOUNT")
plt.title("SCATTERR PLOT")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
x=['sun','mon','tues','wed','thu','fri','sat']
y=[10,7,4,6,2,5,3]
sns.pointplot(x=x,y=y,palette="rocket",markers='*')
plt.xlabel("DAYS")
plt.ylabel("AMOUNT")
plt.title("POINT PLOT",fontsize=10,fontweight="bold")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
x=['sun','mon','tues','wed','thu','fri','sat']
y=[10,7,4,6,2,500,3]
sns.boxplot(x=x,y=y,palette="rocket")
plt.xlabel("DAYS")
plt.ylabel("AMOUNT")
plt.title("BOX PLOT")
plt.show()
```
