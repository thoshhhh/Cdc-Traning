```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from xgboost import XGBRegressor
```

```python
data=pd.read_csv('dataset2.csv')
data
```

```python
data.describe()
```

```python
#step 3 sep dep and independent varaibles
x=data.drop('PRICE',axis=1)
y=data['PRICE']
```

```python
x
```

```python
#splittimg the dataset
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print('80% traning data is (rows)',xtrain.shape)
print('20% testing data is (rows)',xtest.shape)
print("80% training data is ",ytrain.shape)
print("80% training data is ",ytest.shape)

```

```python
#step 5 load the algoritm which u want to use
from xgboost import XGBRegressor
#step 6 to train the function
xgb=XGBRegressor()
xgb.fit(xtrain,ytrain)
```

```python
#step 7 prediction stage for testing
output = xgb.predict(xtest)
print(output)
```

```python
print(xtest)
```

```python
print(abs(r2_score(output,ytest)))
```
